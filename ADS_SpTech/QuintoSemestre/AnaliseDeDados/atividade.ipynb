```python
# 1.1) Cadastro básico - Crie variáveis para nome e idade. Mostre o tipo de cada variável e calcule o ano
# aproximado de nascimento considerando 2025.

from datetime import date

nome = "Murilo"
idade = 19

print("Tipo da variável nome: ", type(nome))
print("Tipo da variável idade: ", type(idade))

print("Ano aproximado de nascimento: ", date.today().year - idade)
```

```python
# 1.2) Altura - Crie uma variável de altura em metros, mostre o tipo e converta para centímetros.
altura = 1.8
print("Tipo da variável altura: ", type(altura))
print("Altura em centímetros: ", altura * 100)
```

```python
# 1.3) Documento - Crie um documento como texto, gere uma versão numérica e some 100 ao valor
texto = "200"
numero = int(texto)
print("Número: ", numero)
print("Número somado com 100: ", numero + 100)
```

```python
# 1.4) Quantidade de caixas – Entre com a quantidade de caixas e calcule o total de itens (12 por caixa).
caixas = int(input("Quantidade de caixas: "))
print("Total de itens: ", caixas * 12)
```

```python
# 1.5) Preço com desconto - Entre o preço do produto, calcule 10% de desconto e mostre o valor final.
preco = float(input("Preço do produto: "))
print("Valor final: ", preco - (preco * 0.10))
```

```python
# 1.6) Solicita dois valores ao usuário e calcula a soma entre eles
a = float(input("Digite o primeiro valor: "))
b = float(input("Digite o segundo valor: "))

soma = a + b
print("Soma: ", soma)
```
